```python
df_test = df_test.drop('Usage', axis=1)
```

```python
df_train = df_train.drop(['Id', 'customer_id'], axis=1)
df_test = df_test.drop(['Id', 'customer_id'], axis=1)
```

```python
df_train = df_train.drop(['csat_scores.appointment', 'csat_scores.email', 'csat_scores.phone', 'csat_scores.whatsapp'], axis=1)
df_test = df_test.drop(['csat_scores.appointment', 'csat_scores.email', 'csat_scores.phone', 'csat_scores.whatsapp'], axis=1)
```

```python
from sklearn.preprocessing import  LabelEncoder
```

```python
for i in df_train.columns:
    if df_train[i].dtype == object:
        try:
            label = LabelEncoder()
            df_train[i] = label.fit_transform(df_train[i])
            df_test[i] = label.transform(df_test[i])
        except:
            print(i)
```

```python
df_train = df_train.drop(['date_of_birth', 'address', 'touchpoints'], axis=1)
df_test = df_test.drop(['date_of_birth', 'address', 'touchpoints'], axis=1)
```

```python
df_train = df_train.drop('date', axis=1)
df_test = df_test.drop('date', axis=1)
```

```python
df_train = df_train.drop('name', axis=1)
df_test = df_test.drop('name', axis=1)
```

```python
X_train = df_train.drop('churn_due_to_fraud', axis=1)
y_train = df_train['churn_due_to_fraud']
X_test = df_test.drop('churn_due_to_fraud', axis=1)
y_test = df_test['churn_due_to_fraud']
```

```python
X_train = X_train.astype(float)
X_test = X_test.astype(float)
```

```python
# X_train = X_train.values.reshape(3926058, 1, 17)
# X_test = X_test.values.reshape(1360472, 1, 17)
```

```python
from keras.utils import to_categorical
```

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(32, activation='relu', input_shape=(17, )))
model.add(Dense(16, activation='relu'))

model.add(Dense(2, activation='softmax'))
```

```python
model.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test), batch_size=32)
```

```python
model.get_weights()
```

```python
y_pred = model.predict(X_test)
```

```python
df_sub['churn'] = y_pred
```

```python
df_sub.to_csv('sub.csv', index=False, index_label=False)
```
